```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython import display
from IPython.display import clear_output
```

```python
file_list = os.listdir('/hdd_c/data/miniWorld/log/')
plt.figure(figsize=(4,4))
for file_name in file_list:
    print(file_name)
    if file_name.endswith('.npz'):
        data = np.load('/hdd_c/data/miniWorld/log/'+file_name)

        s=data['step']
        r=data['reward']

        if len(s)>100:
            print(s.shape)
            print(r.shape)
            plt.plot(s,r)
        
#plt.title('Purely symbolic')
plt.xlabel('Time steps')
plt.ylabel('Average reward')
plt.show()
```
